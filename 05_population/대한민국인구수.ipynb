```python
import pandas as pd
import re
df = pd.read_csv("대한민국인구10년통계.csv",encoding='cp949',header=1)

pattern = r'\([^)]*\)+.*' 
year = 2012
newColName = []

for e,word in enumerate(df.columns) :
    text = re.sub(pattern=pattern, repl='',string=word)
    
    if e != 0 :
        text = str(year) + ' ' + text
    
    if e %3 == 0 :
        year += 1
    
    newColName.append(text)
```

```python
df.columns = newColName
df.info()
```
